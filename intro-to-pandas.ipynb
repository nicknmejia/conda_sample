```python
# Column from series
seats_column = pd.Series([5,5,5,5,5])

# New column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# Column from list
# When made from list, must be same length as columns
fuel_economy = [7.5,9.2,5.0,9.6,8.7,7.5,9.2,5.0,9.6,8.7]
car_sales["Economy"] = fuel_economy
```

```python
car_sales
```

```python
car_sales["Usage"] = car_sales["Odometer (KM)"]/100 * car_sales.Economy
```

```python
car_sales
```

```python
# Create a column from a single value
car_sales["Wheels"] = 4
```

```python
car_sales
```
